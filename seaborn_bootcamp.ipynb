```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,6))
```

```python
#1 - Scatter Plot
sns.scatterplot(data=tips, x='total_bill', y='tip', hue="time", size="size", palette='deep')
plt.title('Total Bill vs Tip')
plt.show()
```

```python
#2 - Line Plot
plt.figure(figsize=(8,6))
sns.lineplot(data=tips, x='size', y='total_bill', hue='sex', marker='o')
plt.title('Line Plot of Total Bill by Size and Sex')
plt.show()
```

```python
#3 - Bar Plot
sns.barplot(data=tips, x='day', y='total_bill', hue='sex', palette='muted')
plt.title('Average Total Bill by Day and Sex')
plt.show()
```

```python
#4 - Box Plot
sns.boxplot(data=tips, x='day', y='tip', hue='smoker', palette='Set2')
plt.title('Box Plot of Tips by Day and Smoking Status')
plt.show()
```

```python
#5 - violin Plot
sns.violinplot(data=tips, x='day', y='total_bill', hue='time', split=True, palette='Pastel1')
plt.title('Violin Plot of Total Bill by Day and Time')
plt.show()
```

```python
tips.columns
```

```python
#6 - count Plot
sns.countplot(data=tips, x='day', hue='smoker', palette='dark')
plt.title('Count of Customers by Day and Smoking Status')
plt.show()
```

```python
#7 - regression Plot
sns.regplot(data=tips, x='total_bill', y='tip', scatter_kws={'s':50}, line_kws={'color':'red'})
plt.title('Regression Plot of Total Bill vs Tip')
plt.show()
```

```python
#8 - histogram
sns.histplot(data=tips, x='total_bill', bins=20, kde=True, color='skyblue')
plt.title('Histogram of Total Bill with KDE')
plt.show()
```

```python
#9 - pair Plot
sns.pairplot(tips, hue='sex', vars=['total_bill', 'tip', 'size'], palette='husl')
plt.suptitle("pair Plot of Total Bill, Tip, and Size by Sex", y=1.02)
plt.show()
```

```python
# 10 - catalog Plot
sns.catplot(data=tips, x='day', y='tip', hue='sex', kind='point', palette='bright')
plt.title('Catalog Plot of Tips by Day and gender')
plt.show()
```

```python
# 11 - joint Plot
sns.jointplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='sex', palette='coolwarm')
plt.suptitle('Joint Plot of Total Bill vs Tip', y=1.02)
plt.show()
```

```python
#12 - Facet Grid
g = sns.FacetGrid(tips, col='time', row='smoker', margin_titles=True).map(sns.histplot, 'total_bill', bins=15, color='purple')
g
```

```python
# 13 - strip Plot
sns.stripplot(data=tips, x='day', y='total_bill', hue='smoker', jitter=True, palette='Set1')
plt.title('Strip Plot of Total Bill by Day and Smoking Status')
plt.show()
```

```python
# 14 - KDE Plot
sns.kdeplot(data=tips, x='total_bill', hue='sex', fill=True, palette='tab10')
plt.title('KDE Plot of Total Bill by Time')
plt.show()
```
